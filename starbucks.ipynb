```python
starbucks_list = []

for item in starbucks_soup_list:

    name = item('strong')[0].text.strip()
    lat= item['data-lat']
    long= item['data-long']
    starbucks_list.append([name,lat,long])


print(starbucks_list)
```

```python
starbucks_list = []

for item in starbucks_soup_list:

    name = item.select('strong')[0].text.strip()
    lat= item['data-lat']
    long= item['data-long']
    address = str(item.select('p')[0]).split('<br/>')[0].split('>')[1]
    tel = str(item.select('p')[0]).split('<br/>')[1].split('<')[0]
    store_type = item.select('i')[0]['class'][0][4:]
    starbucks_list.append([name,lat,long,store_type,address,tel])


print(starbucks_list)
```
